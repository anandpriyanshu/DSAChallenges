```python
from Tkinter import Frame, Label, Center    # tkinter is a grid managr

import LogicsFinal
import constants as c
# till only create the frame and nothing is added till this point
class Gams2048(Frame):    # Inherit Frame classs in game 
    def __init__(self):     
        Frame.__init__(self)      # Frame initiliaze
        self.grid()                  # visualize frame as a grid
        self.master.title('2048')      
        self.master.bind('<Key>', self.key_down)     # press any key bind with keydown function
        self.commands = {c.KEY_UP : LogicsFinal.move_up, c.KEY_DOWN:LogicsFinal.move_down, 
                         c.KEY_LEFT:LogicsFinal.move_left,c.KEY_RIGHT:LogicsFinal.move_right}
        
        # added in the frame by using some function
        self.grid_cells = []     # intially grid is empty
        self.init_grid()         # intialize the grid
        self.init_matrix()       # intilaize the matrix functionb means add 2 in matrix
        self.update_grid_cells()   # maintain the matrix accoring to matrix make changes into UI
        self.mainloop()               # now we can start the game 
        
    def init_grid(self):
        background = Frame(self, bg = c.BACKGROUND_COLOR_GAME,    # create frame inside frame
                           width = c.SIZE,height = c.SIZE)
        background.grid()    # frame as a grid
        
        for i in range(c.GRID_LEN):
            grid_row = []   
            for j in range(c.GRID_LEN):
                cell = Frame(backgroundground,bg = c.BACKGROUND_COLOR_CELL_EMPTY,      #inside frame there is cell
                            width = c.SIZE/GRID_LEN,
                            height = c.SIZE/GRID_LEN)
                cell.grid(row=i,column = j, padx = c.GRID_PADDING,     # cell visualize as grid and cell with padding 
                          pady = c.GRID_PADDING)
                t = Label(master = cell,text = '',             # inside cell there is a lebel i =n which texts are present
                          bg = c.BACKGROUND_COLOR_CELL_EMPTY,
                         justify = CENTER, font = c.FONT,width=5, height=2)
                t.grid()           # lebel as a grid
                grid_row.append(t)               # all the lebel are appended inti the grid row like [l1,l2,l3.....l16]
            self.grid_cells.append(grid_row)  # append grid rows in grid cells like [[l1,l2l3,l4],[l5,l6,l7,l8]........l16]
            
    def init_matrix(self):
        self.matrix = LogicsFinal.start_game()
        LogicsFinal.add_new_2(self.matrix)
        LogicsFinal.add_new_2(self.matrix)
    
    # maintain the matrix accoring to matrix make changes in UI
    def update_grid_cells(self):
        for i in range(c.GRID_LEN):
            for j in range(c.GRID_LEN):
                new_number = self.matrix[i][j]
                if new_number == 0:   # if new_number is empty then text is empty in cell and their respective color
                    self.grid_cells[i][j].configure(
                    text='',bg = c.BACKGROUND_COLOR_CELL_EMPTY)
                else:
                    self.grid_cells[i][j].configure(text = str(new_number),    # if new_number is not empty then text is empty in cell and their respective color
                    bg = c.BACKGROUND_COLOR_DICT[new_number],
                    fg = c.CELL_COLOR_DICT[new_number])
        self.update_idletasks()      # it's frame library, it will wait until all the color change
         
```

```python
import random
def start_game():
    mat = []
    for i in range(4):
        mat.append([0]*4)
    return mat

def generaterandom(grid, emptycells):
    cell = random.choice(emptycells)
    li = [2,4]
    val = random .choice(li)
    i = cell[0]
    j = cell[1]
    grid[i][j] = val

def checkempty(grid):
    cells = []
    for i in range(4):
        for j in range(4):
            if grid[i][j] == 0:
                cells.append([i,j])
    
    return cells

def move_up(grid):
    change = False
    for j in range(0,4):
        top = [0,j]
        for i in range(1,4):
            current = grid[i][j]
            
            if current == 0:
                continue
            
            elif current == grid[top[0]][top[1]]:
                grid[top[0]][top[1]] *= 2
                grid[i][j] = 0
                top = [top[0]+1,top[1]]
                change = True
                
            elif grid[top[0]][top[1]] == 0:
                grid[top[0]][top[1]] = current
                grid[i][j] = 0
                change = True
            
            else:
                grid[top[0]+1][top[1]] = current
                
                if top[0]+1 != i:
                    grid[i][j] = 0
                    change = True
                
                top = [top[0]+1,top[1]]
    
    return grid

def move_down(grid):
    change = False
    for j in range(0,4):
        bottom = [3,j]
        i = 2
        while i > -1:
            current = grid[i][j]
            
            if current == 0:
                i -= 1
                continue
            
            elif current == grid[bottom[0]][bottom[1]]:
                grid[bottom[0]][bottom[1]] *= 2
                grid[i][j] = 0
                bottom = [bottom[0]-1,bottom[1]]
                change = True
                
            elif grid[bottom[0]][bottom[1]] == 0:
                grid[bottom[0]][bottom[1]] = current
                grid[i][j] = 0
                change = True
            
            else:
                grid[bottom[0]-1][bottom[1]] = current
                
                if bottom[0]-1 != i:
                    change = True
                    grid[i][j] = 0
                
                bottom = [bottom[0]-1,bottom[1]]
            
            i -= 1
    
    return grid

def move_right(grid):
    change = False
    for i in range(0,4):
        right = [i,3]
        j = 2
        while j > -1:
            current = grid[i][j]
            
            if current == 0:
                j -= 1
                continue
            
            elif current == grid[right[0]][right[1]]:
                grid[right[0]][right[1]] *= 2
                grid[i][j] = 0
                right = [right[0],right[1]-1]
                change = True
                
            elif grid[right[0]][right[1]] == 0:
                grid[right[0]][right[1]] = current
                grid[i][j] = 0
                change = True
            
            else:
                grid[right[0]][right[1]-1] = current
                
                if right[1]-1 != j:
                    change = True
                    grid[i][j] = 0
                
                right = [right[0],right[1]-1]
            
            j -= 1
    
    return grid

def move_left(grid):
    change = False
    for i in range(4):
        left = [i,0]
        for j in range(1,4):
            current = grid[i][j]
                
            if current == 0:
                continue
                
            elif current == grid[left[0]][left[1]]:
                grid[left[0]][left[1]] *= 2
                grid[i][j] = 0
                left[1] += 1
                change = True
                
            elif grid[left[0]][left[1]] == 0:
                grid[left[0]][left[1]] = current
                grid[i][j] = 0
                change = True
                
            else:
                grid[left[0]][left[1]+1] = current
                    
                if left[1]+1 != j:
                    grid[i][j] = 0
                    change = True
                    
                left[1] += 1
    
    return grid


mat = start_game()
mat[1][3] = 2
mat[2][2] = 2
mat[3][0] = 4
mat[3][1] = 8
mat[2][1] = 4
inputs = [int(ele) for ele in input().split()]
for ele in inputs:
    if ele == 1:
        mat = move_up(mat)
    elif ele == 2:
        mat = move_down(mat)
    elif ele == 3:
        mat = move_left(mat)
    else:
        mat = move_right(mat)
    print(mat)
```
