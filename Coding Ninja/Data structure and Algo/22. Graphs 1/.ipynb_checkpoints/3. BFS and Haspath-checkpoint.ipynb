```python
import queue
class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjMatrix=[[0 for i in range(nVertices)]for j in range(nVertices)]
        
    def addEdge(self,v1,v2):
        self.adjMatrix[v1][v2]=1
        self.adjMatrix[v2][v1]=1
        
    def __DFSHelper(self,sv,visited):                          #sv =, source vertices
        print(sv)                                               #print sv and mark visIted
        visited[sv]=True
        for i in range(self.nVertices):
            if self.adjMatrix[sv][i]>0 and visited[i] is False:  #if there is an edge  b/w sv and i  and not visited
                self.__DFSHelper(i,visited)                      # start sv from i which is not visIted
                
        
        
    def DFS(self):
        visited=[False for i in range(self.nVertices)]    #initially all are not visited
        self.__DFSHelper(0,visited)                       # start with zero and not visited
        
    def BFS(self):
        q = queue.Queue()                         #create queue
        q.put(0)                           #initially put first element(like zero) in q   
        visited=[False for i in range(self.nVertices)]
        visited[0] = True         
        
        while q.empty() is False:
            u = q.get()                 #come out of the queue
            print(u)                    #print u means current element
            for i in range(self.nVertices):
                if self.adjMatrix[u][i]>0 and visited[i] is False:
                    q.put(i)          # put(append) the i in queue
                    visited[i] = True
                    
    def removeEdge(self):
        if self.containsEdge(v1,v2) is False:
            return
        self.adjMatrix[v1][v2]=0
        self.adjMatrix[v2][v1]=0
        
    def containsEdge(self,v1,v2):
        return True if self.adjMatrix[v1][v2]>0 else False
    
    def __str__(self):
        return str(self.adjMatrix)
g=Graph(5)
g.addEdge(0,1)
g.addEdge(1,3)
g.addEdge(2,4)
g.addEdge(0,2)
g.BFS()
```

```python
#TA solution

import sys
sys.setrecursionlimit(10**6)

class graph:
    
    def __init__(self, vertices):
        self.matrix = [[0 for j in range(vertices)] for i in range(vertices)]
        self.vertices = vertices
    
    def addedge(self, v1, v2):
        self.matrix[v1][v2] = 1
        self.matrix[v2][v1] = 1
    
    def __haspath(self, v1, v2, visited):
        if v1 == v2:   #v1 & b2 are adjacent
            return True
        
        visited[v1] = True
        for i in range(self.vertices):
            if self.matrix[v1][i] == 1 and visited[i] == False:
                if self.__haspath(i, v2, visited):
                    return True
        
        return False
    
    def haspath(self, v1, v2):
        visited = [False for i in range(self.vertices)]
        return self.__haspath(v1, v2, visited)

# main
ve = sys.stdin.readline().strip().split(' ')
edges = int(ve[1])
g1 = graph(int(ve[0]))

while edges > 0:
    vertices = sys.stdin.readline().strip().split(' ')
    g1.addedge(int(vertices[0]), int(vertices[1]))
    
    edges -= 1

vertices = sys.stdin.readline().strip().split(' ')

ans = g1.haspath(int(vertices[0]), int(vertices[1]))
if ans:
    print('true')
else:
    print('false')
```
