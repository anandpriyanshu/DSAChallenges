```python
class BinaryTree:
    
    def __init__(self,data):
        
        self.data =data
        self.left = None
        self.right = None
        
def PrintB(root):
    
    if root is None:
        return
    print(root.data,end=" ")
    PrintB(root.left)
    PrintB(root.right)
    
    
def printBdetail(root):
    if root == None:
        return
    
    print(root.data, end=":")
    if root.left != None:
        print("L", root.left.data, end=",")
    if root.right != None:
        print("R", root.right.data,end="")
    print()
    printBdetail(root.left)
    printBdetail(root.right)
    
def treeInput():
    rootData=int(input())
    if rootData==-1:
        return None
    
    root=BinaryTree(rootData)
    leftTree=treeInput()
    rightTree=treeInput()
    root.left=leftTree
    root.right=rightTree
    return root

def LargestData(root):
    if root == None:
        return 0
    
    leftData = LargestData(root.left)
    rightData = LargestData(root.right)
    Largest = max(leftData,rightData,root.data)
    
    return Largest

def heightMax(root):  #(max_depth)
    
    if root == None:
        return 0
   
    leftH = heightMax(root.left)
    rightH = heightMax(root.right)
    
    maxheight = max(leftH,rightH)
    
    return 1 + maxheight 

def heightMin(root):  #(min_depth)
    
    if root == None:
        return 0
    
    leftD  = heightMin(root.left)
    rightD = heightMin(root.right)
    
    if root.left == None and root.right == None:
        return 1
    
    if root.left == None:
        return 1 + rightD
    if root.right == None:
        return 1 + leftD
    
    return 1 + min(leftD,rightD) 
    
    
def numOfNodes(root):
    
    if root == None:
        return 0
    
    leftcount = numOfNodes(root.left)
    rightcount = numOfNodes(root.right)
    
    return 1 + leftcount + rightcount

def numOfLeafNode(root):
    
    if root == None:
        return 0
    
    if root.left == None and root.right == None:
        return 1
    
    left = numOfLeafNode(root.left)
    right = numOfLeafNode(root.right)
    
    return left + right

def printLeafNode(root):
    
    if root  == None:
        return 0
    
    if root.left == None and root.right == None:
        print(root.data, end = " ")
        
    left = printLeafNode(root.left)
    right = printLeafNode(root.right)
    return root


def IsNodePresent(root,x):
    
    if root  == None:
        return False
    
    if root.data == x:
        return True
    
    Lnode = IsNodePresent(root.left,x)
    
    if Lnode:
        return True
    Rnode = IsNodePresent(root.right,x)
    
    if Rnode:
        return True
    
    else:
        return False

    
def NodeWithoutSibling(root):
        
    if root == None:
        return 
        
    if root.left != None and root.right == None:
        print(root.left.data,end ='')
    if root.right != None and root.left == None:
        print(root.right.data)
    left = NodeWithoutSibling(root.left)
    right = NodeWithoutSibling(root.right)
        

def PrintNodeAtDepthK(root,k):
    if root == None:
        return
    
    if k == 0:
        print(root.data)
        return
    
    
    left = PrintNodeAtDepthK(root.left,k-1)
    right =  PrintNodeAtDepthK(root.right,k-1)
    
    
def RemoveLeaves(root):
    
    if root == None:
        return None
    
    if root.left == None and root.right == None:
        return None
    
    RemoveLeaves(root.left)
    RemoveLeaves(root.right)
    
        
# btn1 = BinaryTree(1)
# btn2 = BinaryTree(2)
# btn3 = BinaryTree(3)

# btn1.left  = btn2
# btn1.right  = btn3
# print(btn1)
# print(btn1.left)
# print(btn1.right)




```
