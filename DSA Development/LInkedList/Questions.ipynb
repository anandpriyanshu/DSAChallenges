### InsertAtIthePosition

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
def printLL(head):
    while head is not None:
        print(head.data,end="->")
        head = head.next
    print("end")
    return


def insertAtIthPosition(head,i,data):
    if i < 0 :
        return head
    
    prev = None
    curr = head
    count = 0
    
    while count < i:
        prev = curr
        curr = curr.next
        count +=1
    newNode = Node(data)
    if prev is not None:
        prev.next  = newNode
    else:
        newNode = head
    newNode.next = curr
    return head
def takeInput():
    
    inputList=[int (ele) for ele in input().split()]
    head = None
    tail = None
    for currData in inputList:
        
        if currData == -1:
            break
            
        newNode = Node(currData)
        
        if head is None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
            
        

    return head
head = takeInput()
printLL(head)
head=insertAtIthPosition(head,2,6)
printLL(head)
```

### RemoveDuplicates

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
def printLL(head):
    while head is not None:
        print(head.data,end="->")
        head = head.next
    print("end")
    return


def RemoveDuplicate(head):
    if head is None or head.next is None:
        return head
    
    curr = head
    while curr.next is not None:
        if curr.data == curr.next.data:
            curr.next = curr.next.next
            
        else:
            curr = curr.next
    return head
    

def takeInput():
    
    inputList=[int (ele) for ele in input().split()]
    head = None
    tail = None
    for currData in inputList:
        
        if currData == -1:
            break
            
        newNode = Node(currData)
        
        if head is None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
            
        

    return head
head = takeInput()
printLL(head)
RemoveDuplicate(head)
printLL(head)
```

### ReverseOfLL

### MidPointOfLL

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None

def MidPointOfLL(head):
    
    s = head
    f = head
    
    while f != None and f.next != None:
        s = s.next
        f = f.next.next
        
    return s

  
def printLL(head):
    while head is not None:
        print(head.data,end="->")
        head = head.next
    print("end")
    return


def takeInput():
    
    inputList=[int (ele) for ele in input().split()]
    head = None
    tail = None
    for currData in inputList:
        
        if currData == -1:
            break
            
        newNode = Node(currData)
        
        if head is None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
            
        

    return head
head=takeInput()
printLL(head)

head = MidPointOfLL(head)
printLL(head)
```
